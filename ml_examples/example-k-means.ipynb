```python
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt
```

### Iris data
* Three species of iris: setosa, vesicolor, verginica
* Four features: sepal length, sepal width, petal length, petal width

```python
iris_data = load_iris()
print(iris_data.data.shape)
print(iris_data.target_names)
print(iris_data.feature_names)
```

### k-means clustering

```python
model = KMeans(n_clusters=3)
model.fit(X=iris_data.data)
```

### Three Centroids

```python
plt.scatter(iris_data.data[:,0][iris_data.target==0],iris_data.data[:,1][iris_data.target==0], color='C0', label= 'setosa')
plt.scatter(iris_data.data[:,0][iris_data.target==1],iris_data.data[:,1][iris_data.target==1], color='C1', label= 'versicolor')
plt.scatter(iris_data.data[:,0][iris_data.target==2],iris_data.data[:,1][iris_data.target==2], color='C2', label= 'verginica')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],marker='*', color='black', s=200, label='cluster centres')
plt.legend()
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.show()
```

```python
plt.scatter(iris_data.data[:,2][iris_data.target==0],iris_data.data[:,3][iris_data.target==0], color='C0', label= 'setosa')
plt.scatter(iris_data.data[:,2][iris_data.target==1],iris_data.data[:,3][iris_data.target==1], color='C1', label= 'versicolor')
plt.scatter(iris_data.data[:,2][iris_data.target==2],iris_data.data[:,3][iris_data.target==2], color='C2', label= 'verginica')
plt.scatter(model.cluster_centers_[:,2],model.cluster_centers_[:,3],marker='*', color='black', s=200, label='cluster centres')
plt.legend()
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.show()
```
